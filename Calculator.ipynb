```python
def add(a, b):
    return a + b
    
print("5 + 3 =", add(5, 3))
```

```python
def subtract(a, b):
    return a - b

print("10 - 4 =", subtract(10, 4))
```

```python
def multiply(a, b):
    return a * b

print("6 Ã— 7 =", multiply(6, 7))
```

```python
def divide(a, b):
    if b == 0:
        return "Error: Zero se divide nahi ho sakta!"
    return a / b

print("20 Ã· 4 =", divide(20, 4))
print("10 Ã· 0 =", divide(10, 0))
```

```python
def power(a, b):
    return a ** b

print("2 ^ 5 =", power(2, 5))
```

```python
def sqrt(a):
    if a < 0:
        return "Error: Negative number!"
    return a ** 0.5

print("âˆš144 =", sqrt(144))
print("âˆš81 =", sqrt(81))
```

```python
def modulus(a, b):
    return a % b

print("17 % 5 =", modulus(17, 5))
```
